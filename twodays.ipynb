```python
counts = {}
with open("./data/radishsurvey.txt") as file:
    for line in file:
        line = line.strip()
        name, vote = line.split(" - ")
        if vote not in counts:
            # First vote for this variety
            counts[vote] = 1
        else:
            # Increment the vote count
            counts[vote] = counts[vote] + 1
print(counts)
```
